````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ADVANCED DATA CLEANING & IMPUTATION
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# Create a copy for data cleaning
df_clean = df.copy()

# Strategy 1: Outlier Detection for Engagement Metrics
engagement_cols = ['engagement_reaction_count', 'engagement_comment_count', 
                   'engagement_share_count', 'engagement_comment_plugin_count']

print("=" * 70)
print("ğŸ“Š OUTLIER ANALYSIS FOR ENGAGEMENT METRICS")
print("=" * 70)

for col in engagement_cols:
    if col in df_clean.columns:
        # Calculate IQR
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 3 * IQR  # Using 3*IQR for extreme outliers
        upper_bound = Q3 + 3 * IQR
        
        # Count outliers
        outliers = df_clean[(df_clean[col] < lower_bound) | (df_clean[col] > upper_bound)][col]
        print(f"{col}:")
        print(f"  Outliers: {len(outliers)} ({len(outliers)/len(df_clean)*100:.2f}%)")
        print(f"  Range: [{df_clean[col].min():.0f}, {df_clean[col].max():.0f}]")
        print(f"  IQR Range: [{lower_bound:.0f}, {upper_bound:.0f}]")
        
        # Cap extreme outliers instead of removing them (preserve data)
        df_clean[col] = df_clean[col].clip(lower=0, upper=upper_bound)

print("=" * 70)

# Strategy 2: Fill engagement metrics with 0 (assuming no engagement if missing)
for col in engagement_cols:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].fillna(0)

# Strategy 3: Fill text fields with placeholder
text_cols = ['title', 'description', 'content', 'author']
for col in text_cols:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].fillna('[MISSING]')

# Strategy 4: Create missing data flags (useful features)
for col in ['content', 'author', 'url_to_image']:
    if col in df.columns and df[col].isnull().sum() > 0:
        df_clean[f'{col}_is_missing'] = df[col].isnull().astype(int)

# Strategy 5: Validate data types
df_clean['top_article'] = df_clean['top_article'].astype(int)

print("\nâœ… DATA CLEANING COMPLETED")
print("=" * 70)
print(f"ğŸ“Š Remaining missing values: {df_clean.isnull().sum().sum()}")
print(f"ğŸ“Š Final dataset shape: {df_clean.shape}")
print(f"âœ… All engagement outliers capped")
print(f"âœ… Missing value flags created")
print("=" * 70)
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# COMPOSITE ENGAGEMENT FEATURES
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# Create feature engineering dataframe
df_features = df_clean.copy()

# Total engagement
df_features['total_engagement'] = (
    df_features['engagement_reaction_count'] +
    df_features['engagement_comment_count'] +
    df_features['engagement_share_count'] +
    df_features['engagement_comment_plugin_count']
)

# Engagement diversity (entropy-like metric)
def calculate_engagement_diversity(row):
    values = [
        row['engagement_reaction_count'],
        row['engagement_comment_count'],
        row['engagement_share_count'],
        row['engagement_comment_plugin_count']
    ]
    total = sum(values)
    if total == 0:
        return 0
    proportions = [v/total for v in values if v > 0]
    return -sum(p * np.log(p) for p in proportions) if proportions else 0

df_features['engagement_diversity'] = df_features.apply(calculate_engagement_diversity, axis=1)

# Virality score (share-to-reaction ratio)
df_features['virality_score'] = np.where(
    df_features['engagement_reaction_count'] > 0,
    df_features['engagement_share_count'] / df_features['engagement_reaction_count'],
    0
)

# Comment depth proxy
df_features['comment_depth_proxy'] = np.where(
    df_features['engagement_comment_count'] > 0,
    df_features['engagement_comment_plugin_count'] / df_features['engagement_comment_count'],
    0
)

print("=" * 70)
print("âœ… COMPOSITE FEATURES CREATED")
print("=" * 70)
print(f"ğŸ“Š Total Engagement - Mean: {df_features['total_engagement'].mean():.2f}")
print(f"ğŸ¯ Engagement Diversity - Mean: {df_features['engagement_diversity'].mean():.2f}")
print(f"ğŸš€ Virality Score - Mean: {df_features['virality_score'].mean():.2f}")
print(f"ğŸ’¬ Comment Depth - Mean: {df_features['comment_depth_proxy'].mean():.2f}")
print("=" * 70)
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# TEMPORAL FEATURE EXTRACTION
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# Convert published_at to datetime if not already
df_features['published_at'] = pd.to_datetime(df_features['published_at'])

# Extract temporal components
df_features['hour_of_day'] = df_features['published_at'].dt.hour
df_features['day_of_week'] = df_features['published_at'].dt.dayofweek
df_features['month'] = df_features['published_at'].dt.month
df_features['quarter'] = df_features['published_at'].dt.quarter
df_features['day_of_month'] = df_features['published_at'].dt.day

# Create binary features
df_features['is_weekend'] = df_features['day_of_week'].isin([5, 6]).astype(int)
df_features['is_morning'] = df_features['hour_of_day'].between(6, 11).astype(int)
df_features['is_afternoon'] = df_features['hour_of_day'].between(12, 17).astype(int)
df_features['is_evening'] = df_features['hour_of_day'].between(18, 23).astype(int)

# Cyclical encoding for hour and day
df_features['hour_sin'] = np.sin(2 * np.pi * df_features['hour_of_day'] / 24)
df_features['hour_cos'] = np.cos(2 * np.pi * df_features['hour_of_day'] / 24)
df_features['day_sin'] = np.sin(2 * np.pi * df_features['day_of_week'] / 7)
df_features['day_cos'] = np.cos(2 * np.pi * df_features['day_of_week'] / 7)

print("=" * 70)
print("âœ… TEMPORAL FEATURES EXTRACTED")
print("=" * 70)
print(f"ğŸ“… Date Range: {df_features['published_at'].min()} to {df_features['published_at'].max()}")
print(f"â° Hour Range: {df_features['hour_of_day'].min()}-{df_features['hour_of_day'].max()}")
print(f"ğŸ“† Weekend Articles: {df_features['is_weekend'].sum()} ({df_features['is_weekend'].mean()*100:.1f}%)")
print("=" * 70)
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# TEXT STATISTICAL FEATURES
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def extract_text_features(text):
    """Extract comprehensive statistical features from text"""
    if pd.isna(text) or text == '[MISSING]':
        return {
            'char_count': 0,
            'word_count': 0,
            'sentence_count': 0,
            'avg_word_length': 0,
            'punctuation_count': 0,
            'capital_ratio': 0,
            'digit_count': 0
        }
    
    # Basic counts
    char_count = len(text)
    words = text.split()
    word_count = len(words)
    sentence_count = len(sent_tokenize(text)) if text else 0
    
    # Average word length
    avg_word_length = np.mean([len(word) for word in words]) if words else 0
    
    # Punctuation count
    punctuation_count = sum(1 for char in text if char in string.punctuation)
    
    # Capital letter ratio
    capital_count = sum(1 for char in text if char.isupper())
    capital_ratio = capital_count / len(text) if len(text) > 0 else 0
    
    # Digit count
    digit_count = sum(1 for char in text if char.isdigit())
    
    return {
        'char_count': char_count,
        'word_count': word_count,
        'sentence_count': sentence_count,
        'avg_word_length': avg_word_length,
        'punctuation_count': punctuation_count,
        'capital_ratio': capital_ratio,
        'digit_count': digit_count
    }

# Apply to title
title_features = df_features['title'].apply(extract_text_features).apply(pd.Series)
title_features.columns = [f'title_{col}' for col in title_features.columns]

# Apply to description
desc_features = df_features['description'].apply(extract_text_features).apply(pd.Series)
desc_features.columns = [f'desc_{col}' for col in desc_features.columns]

# Combine with main dataframe
df_features = pd.concat([df_features, title_features, desc_features], axis=1)

print("=" * 70)
print("âœ… TEXT STATISTICAL FEATURES EXTRACTED")
print("=" * 70)
print(f"ğŸ“ Title - Avg Word Count: {df_features['title_word_count'].mean():.1f}")
print(f"ğŸ“ Description - Avg Word Count: {df_features['desc_word_count'].mean():.1f}")
print(f"ğŸ“ Title - Avg Char Count: {df_features['title_char_count'].mean():.0f}")
print("=" * 70)
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ENHANCED TEXT FEATURES - URL & DOMAIN ANALYSIS
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

from urllib.parse import urlparse
import pandas as pd

# Extract domain features WITHOUT tldextract
def extract_domain_features(url):
    """Extract domain-related features from URL"""
    if pd.isna(url) or url == '[MISSING]':
        return {
            'domain': 'unknown',
            'subdomain': 'unknown',
            'tld': 'unknown',
            'url_length': 0,
            'has_https': 0
        }
    
    try:
        parsed = urlparse(url)
        netloc = parsed.netloc
        if netloc.startswith('www.'):
            netloc = netloc[4:]
        parts = netloc.split('.')
        if len(parts) >= 2:
            domain = parts[-2]
            tld = parts[-1]
            subdomain = '.'.join(parts[:-2]) if len(parts) > 2 else 'www'
        else:
            domain = netloc
            tld = ''
            subdomain = 'www'
        
        return {
            'domain': domain,
            'subdomain': subdomain,
            'tld': tld,
            'url_length': len(url),
            'has_https': 1 if parsed.scheme == 'https' else 0
        }
    except:
        return {
            'domain': 'unknown',
            'subdomain': 'unknown',
            'tld': 'unknown',
            'url_length': len(str(url)),
            'has_https': 0
        }

# Apply domain extraction
domain_features = df_features['url'].apply(extract_domain_features).apply(pd.Series)
df_features['url_domain'] = domain_features['domain']
df_features['url_length'] = domain_features['url_length']
df_features['url_has_https'] = domain_features['has_https']

# Source credibility (based on domain frequency)
domain_counts = df_features['url_domain'].value_counts()
df_features['source_article_count'] = df_features['url_domain'].map(domain_counts)
df_features['is_frequent_source'] = (df_features['source_article_count'] >= 10).astype(int)

print("=" * 70)
print("âœ… URL & DOMAIN FEATURES EXTRACTED")
print("=" * 70)
print(f"ğŸŒ Unique Domains: {df_features['url_domain'].nunique()}")
print(f"ğŸ”’ HTTPS URLs: {df_features['url_has_https'].sum()} ({df_features['url_has_https'].mean()*100:.1f}%)")
print(f"ğŸ“° Frequent Sources (â‰¥10 articles): {df_features['is_frequent_source'].sum()}")
print("\nğŸ“Š Top 5 Most Frequent Sources:")
print(domain_counts.head())
print("=" * 70)
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# SENTIMENT ANALYSIS FEATURES
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# Initialize VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

def get_sentiment_scores(text):
    """Extract VADER sentiment scores"""
    if pd.isna(text) or text == '[MISSING]':
        return {'neg': 0, 'neu': 1, 'pos': 0, 'compound': 0}
    return sia.polarity_scores(text)

# Title sentiment
title_sentiment = df_features['title'].apply(get_sentiment_scores).apply(pd.Series)
title_sentiment.columns = [f'title_sentiment_{col}' for col in title_sentiment.columns]

# Description sentiment
desc_sentiment = df_features['description'].apply(get_sentiment_scores).apply(pd.Series)
desc_sentiment.columns = [f'desc_sentiment_{col}' for col in desc_sentiment.columns]

# Combine with main dataframe
df_features = pd.concat([df_features, title_sentiment, desc_sentiment], axis=1)

print("=" * 70)
print("âœ… SENTIMENT ANALYSIS COMPLETED")
print("=" * 70)
print(f"ğŸ˜Š Title Positive Sentiment - Avg: {df_features['title_sentiment_pos'].mean():.3f}")
print(f"ğŸ˜ Title Neutral Sentiment - Avg: {df_features['title_sentiment_neu'].mean():.3f}")
print(f"ğŸ˜ Title Negative Sentiment - Avg: {df_features['title_sentiment_neg'].mean():.3f}")
print(f"ğŸ¯ Title Compound Score - Avg: {df_features['title_sentiment_compound'].mean():.3f}")
print("=" * 70)
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ADVANCED TEXT FEATURES - READABILITY & COMPLEXITY
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def calculate_readability_scores(text):
    """Calculate various readability metrics"""
    if pd.isna(text) or text == '[MISSING]' or len(text) == 0:
        return {
            'flesch_reading_ease': 0,
            'avg_syllables_per_word': 0,
            'difficult_words': 0
        }
    
    words = text.split()
    sentences = text.split('.')
    
    if len(words) == 0 or len(sentences) == 0:
        return {
            'flesch_reading_ease': 0,
            'avg_syllables_per_word': 0,
            'difficult_words': 0
        }
    
    # Simple syllable counter
    def count_syllables(word):
        word = word.lower()
        vowels = 'aeiouy'
        count = 0
        prev_char_was_vowel = False
        for char in word:
            is_vowel = char in vowels
            if is_vowel and not prev_char_was_vowel:
                count += 1
            prev_char_was_vowel = is_vowel
        return max(1, count)
    
    # Calculate metrics
    total_syllables = sum(count_syllables(word) for word in words)
    avg_syllables = total_syllables / len(words) if len(words) > 0 else 0
    avg_words_per_sentence = len(words) / len(sentences) if len(sentences) > 0 else 0
    
    # Flesch Reading Ease (simplified)
    flesch_score = 206.835 - 1.015 * avg_words_per_sentence - 84.6 * avg_syllables
    
    # Count difficult words (3+ syllables)
    difficult = sum(1 for word in words if count_syllables(word) >= 3)
    
    return {
        'flesch_reading_ease': max(0, min(100, flesch_score)),
        'avg_syllables_per_word': avg_syllables,
        'difficult_words_ratio': difficult / len(words) if len(words) > 0 else 0
    }

# Apply to title and description
title_readability = df_features['title'].apply(calculate_readability_scores).apply(pd.Series)
df_features['title_flesch_score'] = title_readability['flesch_reading_ease']
df_features['title_complexity'] = title_readability['difficult_words_ratio']

print("=" * 70)
print("âœ… READABILITY FEATURES EXTRACTED")
print("=" * 70)
print(f"ğŸ“– Average Title Flesch Score: {df_features['title_flesch_score'].mean():.2f}")
print(f"   (Higher = Easier to read, 60-70 = Standard, 90-100 = Very Easy)")
print(f"ğŸ“š Average Title Complexity: {df_features['title_complexity'].mean():.2%}")
print("=" * 70)
````

    <h4 style="color: #1565C0; margin: 0;">ğŸ’¡ The Purpose of Model Comparison</h4>
    <br>
    <p><strong>We don't know which algorithm will work best until we try them!</strong></p>
    <br>
    <table style="width: 100%; border-collapse: collapse;">
        <tr style="background-color: #BBDEFB;">
            <th style="padding: 10px; text-align: left; border: 1px solid #90CAF9;">Model</th>
            <th style="padding: 10px; text-align: left; border: 1px solid #90CAF9;">Best For</th>
            <th style="padding: 10px; text-align: left; border: 1px solid #90CAF9;">Weaknesses</th>
        </tr>
        <tr>
            <td style="padding: 8px; border: 1px solid #90CAF9;"><strong>Logistic Regression</strong></td>
            <td style="padding: 8px; border: 1px solid #90CAF9;">âœ… Fast, interpretable, linear relationships</td>
            <td style="padding: 8px; border: 1px solid #90CAF9;">âŒ Can't capture complex patterns</td>
        </tr>
        <tr style="background-color: #E3F2FD;">
            <td style="padding: 8px; border: 1px solid #90CAF9;"><strong>Random Forest</strong></td>
            <td style="padding: 8px; border: 1px solid #90CAF9;">âœ… Handles non-linear data, robust to outliers</td>
            <td style="padding: 8px; border: 1px solid #90CAF9;">âŒ Can be slow, black-box</td>
        </tr>
        <tr>
            <td style="padding: 8px; border: 1px solid #90CAF9;"><strong>XGBoost</strong></td>
            <td style="padding: 8px; border: 1px solid #90CAF9;">âœ… Usually best performance, handles missing data</td>
            <td style="padding: 8px; border: 1px solid #90CAF9;">âŒ Requires tuning, overfitting risk</td>
        </tr>
        <tr style="background-color: #E3F2FD;">
            <td style="padding: 8px; border: 1px solid #90CAF9;"><strong>LightGBM</strong></td>
            <td style="padding: 8px; border: 1px solid #90CAF9;">âœ… Fast training, low memory, large datasets</td>
            <td style="padding: 8px; border: 1px solid #90CAF9;">âŒ Sensitive to overfitting on small data</td>
        </tr>
    </table>
    <br>
    <p><strong>ğŸ¯ Our Strategy:</strong></p>
    <ol>
        <li><strong>Train all models</strong> on the same data</li>
        <li><strong>Compare performance</strong> using validation set</li>
        <li><strong>Select the best</strong> based on F1-Score (not accuracy!)</li>
        <li><strong>Test on real data</strong> to verify it works in production</li>
    </ol>
    <br>
    <p style="background-color: #FFF9C4; padding: 10px; border-radius: 5px; border-left: 4px solid #FBC02D;">
        <strong>âš¡ Key Insight:</strong> For imbalanced data (87% vs 13%), we focus on <strong>F1-Score</strong> and <strong>Recall</strong>, not accuracy. A model that predicts all articles as "regular" would get 87% accuracy but be completely useless!
    </p>
</div>

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# FEATURE SELECTION & PREPARATION
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# Select features for modeling (excluding text columns, URLs, dates, and target)
exclude_cols = ['source_id', 'source_name', 'author', 'title', 'description', 
                'url', 'url_to_image', 'published_at', 'content', 'top_article']

# Get all numeric features
feature_cols = [col for col in df_features.columns 
                if col not in exclude_cols and df_features[col].dtype in [np.float64, np.int64, np.float32, np.int32]]

# Prepare feature matrix and target
X = df_features[feature_cols].fillna(0)  # Fill any remaining NaN with 0

# FIX: Fill NaN in target column before converting to int
y = df_features['top_article'].fillna(0).astype(int)

print("=" * 70)
print("ğŸ“Š FEATURE PREPARATION SUMMARY")
print("=" * 70)
print(f"ğŸ”¢ Total Features: {X.shape[1]}")
print(f"ğŸ“ Feature Names (first 10): {feature_cols[:10]}")
print(f"ğŸ¯ Target Distribution:")
print(y.value_counts())
print(f"ğŸ“Š Feature Matrix Shape: {X.shape}")
print("=" * 70)
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# TRAIN-TEST SPLIT WITH STRATIFICATION
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# First split: Train+Val vs Test
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=Config.TEST_SIZE, random_state=Config.RANDOM_SEED, stratify=y
)

# Second split: Train vs Val
val_size_adjusted = Config.VAL_SIZE / (Config.TRAIN_SIZE + Config.VAL_SIZE)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=val_size_adjusted, 
    random_state=Config.RANDOM_SEED, stratify=y_temp
)

print("=" * 70)
print("âœ… DATA SPLITTING COMPLETED")
print("=" * 70)
print(f"ğŸ“Š Training Set: {X_train.shape[0]:,} samples ({X_train.shape[0]/len(X)*100:.1f}%)")
print(f"ğŸ“Š Validation Set: {X_val.shape[0]:,} samples ({X_val.shape[0]/len(X)*100:.1f}%)")
print(f"ğŸ“Š Test Set: {X_test.shape[0]:,} samples ({X_test.shape[0]/len(X)*100:.1f}%)")
print(f"\nğŸ¯ Class Distribution:")
print(f"   Train: {y_train.value_counts().to_dict()}")
print(f"   Val:   {y_val.value_counts().to_dict()}")
print(f"   Test:  {y_test.value_counts().to_dict()}")
print("=" * 70)
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# HANDLE CLASS IMBALANCE WITH SMOTE
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("=" * 70)
print("ğŸ”„ APPLYING SMOTE FOR CLASS IMBALANCE")
print("=" * 70)

# Check if SMOTE is available
try:
    from imblearn.over_sampling import SMOTE
    
    # Store original distribution
    original_dist = y_train.value_counts()
    print(f"ğŸ“Š Original Training Distribution:")
    print(f"   Class 0 (Regular): {original_dist[0]:,} ({original_dist[0]/len(y_train)*100:.1f}%)")
    print(f"   Class 1 (Top): {original_dist[1]:,} ({original_dist[1]/len(y_train)*100:.1f}%)")
    print(f"   Imbalance Ratio: {original_dist[1]/original_dist[0]:.2f}")
    
    # Apply SMOTE with sampling strategy of 0.5 (minority will be 50% of majority)
    smote = SMOTE(sampling_strategy=0.5, random_state=Config.RANDOM_SEED, k_neighbors=5)
    X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)
    
    # New distribution
    new_dist = pd.Series(y_train_balanced).value_counts()
    print(f"\nğŸ“Š After SMOTE Distribution:")
    print(f"   Class 0 (Regular): {new_dist[0]:,} ({new_dist[0]/len(y_train_balanced)*100:.1f}%)")
    print(f"   Class 1 (Top): {new_dist[1]:,} ({new_dist[1]/len(y_train_balanced)*100:.1f}%)")
    print(f"   Imbalance Ratio: {new_dist[1]/new_dist[0]:.2f}")
    print(f"\nâœ… Training set size: {len(y_train):,} â†’ {len(y_train_balanced):,} samples")
    print(f"âœ… Synthetic samples created: {len(y_train_balanced) - len(y_train):,}")
    
    # Use balanced data for training
    X_train_final = X_train_balanced
    y_train_final = y_train_balanced
    
except ImportError:
    print("âš ï¸  SMOTE not available. Using class weights instead.")
    print("   Install with: pip install imbalanced-learn")
    X_train_final = X_train
    y_train_final = y_train

print("=" * 70)
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# FEATURE SCALING
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# Use RobustScaler to handle outliers better than StandardScaler
scaler = RobustScaler()

# Fit on training data (use balanced data if SMOTE was applied)
try:
    X_train_scaled = scaler.fit_transform(X_train_final)
except NameError:
    # If SMOTE wasn't available
    X_train_final = X_train
    y_train_final = y_train
    X_train_scaled = scaler.fit_transform(X_train)

# Transform validation and test sets
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

print("=" * 70)
print("âœ… FEATURE SCALING COMPLETED (RobustScaler)")
print("=" * 70)
print(f"ï¿½ Scaled Training Shape: {X_train_scaled.shape}")
print(f"ğŸ“Š Scaled Validation Shape: {X_val_scaled.shape}")
print(f"ï¿½ Scaled Test Shape: {X_test_scaled.shape}")
print("\nï¿½ RobustScaler used for better outlier handling")
print("   (Uses median and IQR instead of mean and std)")
print("=" * 70)
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# RANDOM FOREST WITH CLASS WEIGHTS
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("=" * 70)
print("ğŸŒ² TRAINING RANDOM FOREST CLASSIFIER")
print("=" * 70)

# Calculate class weights for imbalanced data
class_weights = 'balanced'  # Automatically adjust weights

rf_model = RandomForestClassifier(
    n_estimators=Config.RF_N_ESTIMATORS,
    max_depth=Config.RF_MAX_DEPTH,
    min_samples_split=Config.RF_MIN_SAMPLES_SPLIT,
    class_weight=class_weights,  # Handle class imbalance
    random_state=Config.RANDOM_SEED,
    n_jobs=-1,
    verbose=1
)

# Train model
rf_model.fit(X_train_scaled, y_train_final)

# Predictions on validation set
y_val_pred = rf_model.predict(X_val_scaled)
y_val_pred_proba = rf_model.predict_proba(X_val_scaled)[:, 1]

# Calculate metrics
rf_accuracy = accuracy_score(y_val, y_val_pred)
rf_precision = precision_score(y_val, y_val_pred, zero_division=0)
rf_recall = recall_score(y_val, y_val_pred)
rf_f1 = f1_score(y_val, y_val_pred)
rf_auc = roc_auc_score(y_val, y_val_pred_proba)

print("\nğŸ“Š RANDOM FOREST PERFORMANCE (Validation Set):")
print("=" * 70)
print(f"ğŸ¯ Accuracy:  {rf_accuracy:.4f}")
print(f"ğŸ¯ Precision: {rf_precision:.4f} (What % of predicted top articles are actually top)")
print(f"ğŸ¯ Recall:    {rf_recall:.4f} (What % of actual top articles were caught)")
print(f"ğŸ¯ F1-Score:  {rf_f1:.4f} (Harmonic mean of precision & recall)")
print(f"ğŸ¯ ROC-AUC:   {rf_auc:.4f} (Overall discrimination ability)")
print("=" * 70)

# Important note for imbalanced data
print("\nğŸ’¡ For imbalanced data, F1-Score and ROC-AUC are more important than Accuracy!")
print("=" * 70)
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# MULTI-MODEL TRAINING & COMPARISON
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("=" * 70)
print("ğŸ¤– TRAINING MULTIPLE MODELS FOR COMPARISON")
print("=" * 70)

# Dictionary to store all models and their results
models = {}
results = []

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# MODEL 1: LOGISTIC REGRESSION
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("\n1ï¸âƒ£ Training Logistic Regression...")
lr_model = LogisticRegression(
    max_iter=1000,
    class_weight='balanced',
    random_state=Config.RANDOM_SEED,
    n_jobs=-1
)
lr_model.fit(X_train_scaled, y_train_final)
y_val_pred_lr = lr_model.predict(X_val_scaled)
y_val_proba_lr = lr_model.predict_proba(X_val_scaled)[:, 1]

lr_metrics = {
    'Model': 'Logistic Regression',
    'Accuracy': accuracy_score(y_val, y_val_pred_lr),
    'Precision': precision_score(y_val, y_val_pred_lr, zero_division=0),
    'Recall': recall_score(y_val, y_val_pred_lr),
    'F1-Score': f1_score(y_val, y_val_pred_lr),
    'ROC-AUC': roc_auc_score(y_val, y_val_proba_lr)
}
results.append(lr_metrics)
models['Logistic Regression'] = lr_model
print(f"âœ… LR - F1: {lr_metrics['F1-Score']:.4f}, AUC: {lr_metrics['ROC-AUC']:.4f}")

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# MODEL 2: RANDOM FOREST
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("\n2ï¸âƒ£ Training Random Forest...")
rf_model_new = RandomForestClassifier(
    n_estimators=Config.RF_N_ESTIMATORS,
    max_depth=Config.RF_MAX_DEPTH,
    min_samples_split=Config.RF_MIN_SAMPLES_SPLIT,
    class_weight='balanced',
    random_state=Config.RANDOM_SEED,
    n_jobs=-1
)
rf_model_new.fit(X_train_scaled, y_train_final)
y_val_pred_rf = rf_model_new.predict(X_val_scaled)
y_val_proba_rf = rf_model_new.predict_proba(X_val_scaled)[:, 1]

rf_metrics = {
    'Model': 'Random Forest',
    'Accuracy': accuracy_score(y_val, y_val_pred_rf),
    'Precision': precision_score(y_val, y_val_pred_rf, zero_division=0),
    'Recall': recall_score(y_val, y_val_pred_rf),
    'F1-Score': f1_score(y_val, y_val_pred_rf),
    'ROC-AUC': roc_auc_score(y_val, y_val_proba_rf)
}
results.append(rf_metrics)
models['Random Forest'] = rf_model_new
print(f"âœ… RF - F1: {rf_metrics['F1-Score']:.4f}, AUC: {rf_metrics['ROC-AUC']:.4f}")

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# MODEL 3: GRADIENT BOOSTING
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("\n3ï¸âƒ£ Training Gradient Boosting...")
gb_model = GradientBoostingClassifier(
    n_estimators=100,
    max_depth=5,
    learning_rate=0.1,
    random_state=Config.RANDOM_SEED
)
gb_model.fit(X_train_scaled, y_train_final)
y_val_pred_gb = gb_model.predict(X_val_scaled)
y_val_proba_gb = gb_model.predict_proba(X_val_scaled)[:, 1]

gb_metrics = {
    'Model': 'Gradient Boosting',
    'Accuracy': accuracy_score(y_val, y_val_pred_gb),
    'Precision': precision_score(y_val, y_val_pred_gb, zero_division=0),
    'Recall': recall_score(y_val, y_val_pred_gb),
    'F1-Score': f1_score(y_val, y_val_pred_gb),
    'ROC-AUC': roc_auc_score(y_val, y_val_proba_gb)
}
results.append(gb_metrics)
models['Gradient Boosting'] = gb_model
print(f"âœ… GB - F1: {gb_metrics['F1-Score']:.4f}, AUC: {gb_metrics['ROC-AUC']:.4f}")

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# MODEL 4: XGBoost (if available)
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
try:
    print("\n4ï¸âƒ£ Training XGBoost...")
    import xgboost as xgb
    
    # Calculate scale_pos_weight for imbalanced data
    scale_pos_weight = (y_train_final == 0).sum() / (y_train_final == 1).sum()
    
    xgb_model = xgb.XGBClassifier(
        n_estimators=Config.XGB_N_ESTIMATORS,
        max_depth=Config.XGB_MAX_DEPTH,
        learning_rate=Config.XGB_LEARNING_RATE,
        scale_pos_weight=scale_pos_weight,
        random_state=Config.RANDOM_SEED,
        n_jobs=-1,
        eval_metric='logloss'
    )
    xgb_model.fit(X_train_scaled, y_train_final)
    y_val_pred_xgb = xgb_model.predict(X_val_scaled)
    y_val_proba_xgb = xgb_model.predict_proba(X_val_scaled)[:, 1]
    
    xgb_metrics = {
        'Model': 'XGBoost',
        'Accuracy': accuracy_score(y_val, y_val_pred_xgb),
        'Precision': precision_score(y_val, y_val_pred_xgb, zero_division=0),
        'Recall': recall_score(y_val, y_val_pred_xgb),
        'F1-Score': f1_score(y_val, y_val_pred_xgb),
        'ROC-AUC': roc_auc_score(y_val, y_val_proba_xgb)
    }
    results.append(xgb_metrics)
    models['XGBoost'] = xgb_model
    print(f"âœ… XGB - F1: {xgb_metrics['F1-Score']:.4f}, AUC: {xgb_metrics['ROC-AUC']:.4f}")
except ImportError:
    print("âš ï¸ XGBoost not available - skipping")

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# CREATE COMPARISON TABLE
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("\n" + "=" * 70)
print("ğŸ“Š MODEL PERFORMANCE COMPARISON")
print("=" * 70)

results_df = pd.DataFrame(results)
results_df = results_df.sort_values('F1-Score', ascending=False)
display(results_df.style.highlight_max(subset=['Precision', 'Recall', 'F1-Score', 'ROC-AUC'], 
                                        color='lightgreen'))

# Select best model
best_model_name = results_df.iloc[0]['Model']
best_f1_score = results_df.iloc[0]['F1-Score']
best_model = models[best_model_name]

print(f"\nï¿½ WINNING MODEL: {best_model_name}")
print(f"ğŸ“ˆ Best F1-Score: {best_f1_score:.4f}")
print("=" * 70)
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CONFUSION MATRIX ANALYSIS
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Calculate confusion matrix
cm = confusion_matrix(y_val, y_val_pred)

fig, ax = plt.subplots(figsize=(8, 6))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, 
                               display_labels=['Regular', 'Top Article'])
disp.plot(cmap='Blues', ax=ax, values_format='d')

ax.set_title('ğŸ¯ Confusion Matrix - Validation Set', fontsize=14, fontweight='bold', pad=15)

# Calculate and display metrics from confusion matrix
tn, fp, fn, tp = cm.ravel()
specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
npv = tn / (tn + fn) if (tn + fn) > 0 else 0  # Negative Predictive Value

# Add text box with additional metrics
metrics_text = f"""
True Negatives: {tn} | False Positives: {fp}
False Negatives: {fn} | True Positives: {tp}

Specificity: {specificity:.3f}
NPV: {npv:.3f}
"""
ax.text(1.5, 0.5, metrics_text, transform=ax.transData,
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5),
        fontsize=10, verticalalignment='center')

plt.tight_layout()
plt.show()

print("=" * 70)
print("ğŸ“Š CONFUSION MATRIX INTERPRETATION")
print("=" * 70)
print(f"True Positives (TP): {tp} - Correctly identified top articles")
print(f"False Positives (FP): {fp} - Regular articles wrongly predicted as top")
print(f"False Negatives (FN): {fn} - Top articles missed by the model")
print(f"True Negatives (TN): {tn} - Correctly identified regular articles")
print(f"\nğŸ’¡ For business: FN is costly (missing top articles)")
print(f"   Current Miss Rate: {fn/(tp+fn)*100:.1f}% of top articles are missed")
print("=" * 70)
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# PRECISION-RECALL CURVE (Better for Imbalanced Data)
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

from sklearn.metrics import precision_recall_curve, average_precision_score

# Calculate precision-recall curve
precision_curve, recall_curve, thresholds_pr = precision_recall_curve(y_val, y_val_pred_proba)
avg_precision = average_precision_score(y_val, y_val_pred_proba)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Precision-Recall Curve
ax1.plot(recall_curve, precision_curve, linewidth=2, color=COLORS['primary'])
ax1.fill_between(recall_curve, precision_curve, alpha=0.2, color=COLORS['primary'])
ax1.set_xlabel('Recall', fontsize=12, fontweight='bold')
ax1.set_ylabel('Precision', fontsize=12, fontweight='bold')
ax1.set_title(f'ï¿½ Precision-Recall Curve\nAP Score: {avg_precision:.3f}', 
             fontsize=14, fontweight='bold', pad=15)
ax1.grid(alpha=0.3, linestyle='--')
ax1.set_xlim([0, 1])
ax1.set_ylim([0, 1])

# Add baseline (random classifier)
baseline = np.sum(y_val) / len(y_val)
ax1.axhline(y=baseline, color='red', linestyle='--', label=f'Baseline: {baseline:.3f}')
ax1.legend()

# Threshold Analysis
f1_scores = 2 * (precision_curve * recall_curve) / (precision_curve + recall_curve + 1e-10)
best_threshold_idx = np.argmax(f1_scores[:-1])  # Exclude last element
best_threshold = thresholds_pr[best_threshold_idx]
best_f1 = f1_scores[best_threshold_idx]

ax2.plot(thresholds_pr, precision_curve[:-1], label='Precision', linewidth=2, color=COLORS['success'])
ax2.plot(thresholds_pr, recall_curve[:-1], label='Recall', linewidth=2, color=COLORS['info'])
ax2.plot(thresholds_pr, f1_scores[:-1], label='F1-Score', linewidth=2, color=COLORS['warning'])
ax2.axvline(x=best_threshold, color='red', linestyle='--', 
           label=f'Optimal: {best_threshold:.3f}')
ax2.set_xlabel('Threshold', fontsize=12, fontweight='bold')
ax2.set_ylabel('Score', fontsize=12, fontweight='bold')
ax2.set_title('ğŸ¯ Threshold Optimization', fontsize=14, fontweight='bold', pad=15)
ax2.legend(loc='best')
ax2.grid(alpha=0.3, linestyle='--')

plt.tight_layout()
plt.show()

print("=" * 70)
print("ğŸ¯ PRECISION-RECALL ANALYSIS")
print("=" * 70)
print(f"ğŸ“Š Average Precision Score: {avg_precision:.4f}")
print(f"ğŸ¯ Optimal Threshold: {best_threshold:.4f}")
print(f"ğŸ“ˆ Best F1-Score at optimal threshold: {best_f1:.4f}")
print(f"\nğŸ’¡ Current model uses threshold: 0.5000")
print(f"ğŸ’¡ Using optimal threshold ({best_threshold:.4f}) could improve F1-Score!")
print("=" * 70)
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# REAL-WORLD ARTICLE PREDICTION FUNCTION (WITH SENTIMENT)
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def predict_article_with_sentiment(title, description="", model=None, scaler_obj=None, features=None):
    """
    Predict if a new article will be a 'Top Article' AND analyze its sentiment.
    
    This is how the model would be used in production!
    
    Parameters:
    -----------
    title : str
        The article headline
    description : str
        Article description/subtitle (optional)
    model : trained model
        The model to use for prediction
    scaler_obj : fitted scaler
        The scaler used during training
    features : list
        List of feature names in correct order
    
    Returns:
    --------
    dict : Prediction results with confidence scores AND sentiment analysis
    """
    
    # Use the best model from training
    if model is None:
        model = rf_model
    if scaler_obj is None:
        scaler_obj = scaler
    if features is None:
        features = feature_cols
    
    # Initialize sentiment analyzer
    try:
        from nltk.sentiment import SentimentIntensityAnalyzer
        sia = SentimentIntensityAnalyzer()
    except:
        print("âš ï¸  NLTK sentiment not available")
        sia = None
    
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    # SENTIMENT ANALYSIS
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    sentiment_data = {}
    
    if sia and title:
        # Analyze title sentiment
        title_sentiment = sia.polarity_scores(title)
        
        # Analyze description sentiment if provided
        if description:
            desc_sentiment = sia.polarity_scores(description)
            combined_text = title + " " + description
            combined_sentiment = sia.polarity_scores(combined_text)
        else:
            desc_sentiment = {'compound': 0, 'pos': 0, 'neg': 0, 'neu': 0}
            combined_sentiment = title_sentiment
        
        # Determine overall sentiment
        compound = combined_sentiment['compound']
        if compound >= 0.05:
            sentiment_label = "ğŸ˜Š Positive"
            sentiment_emoji = "ğŸ˜Š"
        elif compound <= -0.05:
            sentiment_label = "ğŸ˜  Negative"
            sentiment_emoji = "ğŸ˜ "
        else:
            sentiment_label = "ğŸ˜ Neutral"
            sentiment_emoji = "ğŸ˜"
        
        sentiment_data = {
            'overall_sentiment': sentiment_label,
            'sentiment_emoji': sentiment_emoji,
            'compound_score': compound,
            'positive_score': combined_sentiment['pos'],
            'negative_score': combined_sentiment['neg'],
            'neutral_score': combined_sentiment['neu'],
            'title_sentiment': title_sentiment,
            'description_sentiment': desc_sentiment
        }
    else:
        sentiment_data = {
            'overall_sentiment': 'â“ Unknown',
            'sentiment_emoji': 'â“',
            'compound_score': 0,
            'positive_score': 0,
            'negative_score': 0,
            'neutral_score': 0
        }
    
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    # FEATURE EXTRACTION FOR ML MODEL
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    article_features = {}
    
    # Text features
    title_words = title.split() if title else []
    article_features['title_word_count'] = len(title_words)
    article_features['title_char_count'] = len(title)
    article_features['title_avg_word_length'] = np.mean([len(w) for w in title_words]) if title_words else 0
    article_features['title_capital_ratio'] = sum(1 for c in title if c.isupper()) / len(title) if title else 0
    article_features['title_punctuation_count'] = sum(1 for c in title if c in '!?.,;:') if title else 0
    article_features['title_digit_count'] = sum(1 for c in title if c.isdigit()) if title else 0
    article_features['title_has_question'] = int('?' in title) if title else 0
    article_features['title_has_exclamation'] = int('!' in title) if title else 0
    
    # Add sentiment features
    if sia:
        article_features['title_sentiment_compound'] = sentiment_data['title_sentiment']['compound']
        article_features['title_sentiment_pos'] = sentiment_data['title_sentiment']['pos']
        article_features['title_sentiment_neg'] = sentiment_data['title_sentiment']['neg']
        article_features['title_sentiment_neu'] = sentiment_data['title_sentiment']['neu']
    else:
        article_features['title_sentiment_compound'] = 0
        article_features['title_sentiment_pos'] = 0
        article_features['title_sentiment_neg'] = 0
        article_features['title_sentiment_neu'] = 0
    
    # Description features
    if description:
        desc_words = description.split()
        article_features['description_word_count'] = len(desc_words)
        article_features['description_char_count'] = len(description)
    else:
        article_features['description_word_count'] = 0
        article_features['description_char_count'] = 0
    
    # Fill missing features with 0
    for feat in features:
        if feat not in article_features:
            article_features[feat] = 0
    
    # Create feature vector in the correct order
    X_new = np.array([[article_features.get(f, 0) for f in features]])
    
    # Scale features
    X_new_scaled = scaler_obj.transform(X_new)
    
    # Make prediction
    prediction = model.predict(X_new_scaled)[0]
    probabilities = model.predict_proba(X_new_scaled)[0]
    
    # Prepare results
    result = {
        'title': title,
        'description': description,
        'prediction': int(prediction),
        'prediction_label': 'ğŸ”¥ TOP PERFORMING' if prediction == 1 else 'ğŸ“° Regular Article',
        'confidence': float(probabilities[prediction]),
        'probability_regular': float(probabilities[0]),
        'probability_top': float(probabilities[1]),
        'sentiment': sentiment_data,
        'recommendation': ''
    }
    
    # Add business recommendation
    if prediction == 1 and probabilities[1] > 0.7:
        result['recommendation'] = "âœ… HIGHLY RECOMMEND: Feature prominently, promote on social media"
    elif prediction == 1 and probabilities[1] > 0.5:
        result['recommendation'] = "âš ï¸ MODERATE: Consider featuring, monitor performance"
    else:
        result['recommendation'] = "ğŸ“ STANDARD: Publish normally, no special promotion needed"
    
    return result

print("=" * 70)
print("âœ… ENHANCED PREDICTION FUNCTION CREATED (WITH SENTIMENT ANALYSIS)")
print("=" * 70)
print("ğŸ“ Function: predict_article_with_sentiment()")
print("   Input: Article title and description")
print("   Output: Prediction + Sentiment Analysis + Recommendations")
print("=" * 70)
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# PRODUCTION INFERENCE PIPELINE
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

class ArticleEngagementPredictor:
    """
    Production-ready predictor for article engagement.
    Handles feature extraction, preprocessing, and prediction.
    """
    
    def __init__(self, model, scaler, feature_names):
        self.model = model
        self.scaler = scaler
        self.feature_names = feature_names
        self.sia = SentimentIntensityAnalyzer()
    
    def extract_features(self, article_data):
        """Extract features from raw article data"""
        features = {}
        
        # Text features
        title = article_data.get('title', '')
        description = article_data.get('description', '')
        
        # Title features
        title_words = title.split()
        features['title_word_count'] = len(title_words)
        features['title_char_count'] = len(title)
        features['title_avg_word_length'] = np.mean([len(w) for w in title_words]) if title_words else 0
        
        # Sentiment
        title_sentiment = self.sia.polarity_scores(title)
        features['title_sentiment_compound'] = title_sentiment['compound']
        features['title_sentiment_pos'] = title_sentiment['pos']
        features['title_sentiment_neg'] = title_sentiment['neg']
        features['title_sentiment_neu'] = title_sentiment['neu']
        
        # Add other features with defaults
        for feat in self.feature_names:
            if feat not in features:
                features[feat] = 0
        
        return features
    
    def predict(self, article_data):
        """Make prediction for a single article"""
        # Extract features
        features = self.extract_features(article_data)
        
        # Create feature vector in correct order
        X = np.array([[features.get(f, 0) for f in self.feature_names]])
        
        # Scale features
        X_scaled = self.scaler.transform(X)
        
        # Make prediction
        prediction = self.model.predict(X_scaled)[0]
        probability = self.model.predict_proba(X_scaled)[0]
        
        return {
            'prediction': int(prediction),
            'prediction_label': 'Top Article' if prediction == 1 else 'Regular Article',
            'confidence': float(probability[prediction]),
            'probabilities': {
                'regular': float(probability[0]),
                'top_article': float(probability[1])
            }
        }

# Initialize predictor
predictor = ArticleEngagementPredictor(rf_model, scaler, feature_cols)

# Test with a sample article
sample_article = {
    'title': 'Breaking: Major technological breakthrough announced',
    'description': 'Scientists unveil revolutionary discovery that could change everything'
}

result = predictor.predict(sample_article)

print("=" * 70)
print("âœ… PRODUCTION INFERENCE PIPELINE READY")
print("=" * 70)
print("\nğŸ§ª TEST PREDICTION:")
print(f"   Article: {sample_article['title']}")
print(f"   Prediction: {result['prediction_label']}")
print(f"   Confidence: {result['confidence']:.2%}")
print(f"   Probabilities: Regular={result['probabilities']['regular']:.2%}, "
      f"Top={result['probabilities']['top_article']:.2%}")
print("=" * 70)
````
